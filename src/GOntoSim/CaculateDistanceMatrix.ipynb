```python
import GOntoSim
import pandas as pd
from goatools.obo_parser import GODag
from goatools.base import get_godag
go = get_godag("go-basic.obo", optional_attrs={'relationship'})

def calculate_similarity_between_2gene(gene1, gene2):
	# 檢查gene1是否為空集合
	if not gene1 or not gene2:
		return 0.0

	# getting a list of unique GO Terms
	unique_goterms = list(set().union(gene1, gene2))
	
	# Setting the variable S_values
	S_values = [(x,GOntoSim.Semantic_Value(x, go, 'Baseline_LCA_avg')) for x in unique_goterms ]
	S_values = dict(S_values)
	return GOntoSim.Similarity_of_Set_of_GOTerms(gene1, gene2, 'GOntoSim', S_values)

# create similarity matrix by giving a dictionary of Gene Ontology terms and its gene
def create_similarity_matrix(gene_dict):
	# create a dataframe with the index of gene names
	df = pd.DataFrame(index=gene_dict.keys(), columns=gene_dict.keys())
	for gene1 in gene_dict.keys():
		for gene2 in gene_dict.keys():
			print(gene1, gene2)
			if gene1 == gene2:
				df[gene1][gene2] = 1.0
			elif gene1 > gene2:
				df[gene1][gene2] = df[gene2][gene1]
			else:
				df[gene1][gene2] = calculate_similarity_between_2gene(gene_dict[gene1], gene_dict[gene2])
	return df
def create_GONumber_matrix(gene_dict):
	# create a dataframe with the index of gene names
	df = pd.DataFrame(index=gene_dict.keys(), columns=gene_dict.keys())
	for gene1 in gene_dict.keys():
		for gene2 in gene_dict.keys():
			df[gene1][gene2] = len(gene_dict[gene1]) + len(gene_dict[gene2])
	return df
```

```python
# read GOTerm.csv
import ast
df = pd.read_csv('wiki.csv')

# remove rows with empty GO domain and empty GO term accession
df = df[df['GO domain'].notna()]
df = df[df['GO term accession'].notna()]
df
```

```python
# 將GO term accession根據不同 GO domain和 Gene name 創建二維陣列
result_dict = {}
for (a_value, b_value), group in df.groupby(['GO domain', 'Gene name']):
    sub_list = group['GO term accession'].tolist()
    result_dict.setdefault(a_value, {})[b_value] = sub_list
result_dict['molecular_function']

# result_dict['molecular_function'] lengh
print(len(result_dict['molecular_function'])) # it should be 388 genes still debugging
```
