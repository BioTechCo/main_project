### svm with champ normalized data

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
normalized_train = pd.read_csv('../champ_result/all_beta_normalized.csv')
normalized_train
```

```python
normalized_test = pd.read_csv('../champ_result/all_beta_normalized_test.csv')
normalized_test
```

```python
normalized_train_c = normalized_train[normalized_train['Unnamed: 0'].isin(normalized_test['Unnamed: 0'])]
normalized_test_c = normalized_test[normalized_test['Unnamed: 0'].isin(normalized_train['Unnamed: 0'])]
```

```python
print(normalized_train_c.shape)
print(normalized_test_c.shape)
```

```python
id_list = normalized_train_c['Unnamed: 0']
# save id_list to txt
id_list.to_csv('../result/id_list.txt', header=None, index=None, sep=' ', mode='a')
```

```python
X_train = normalized_train_c.iloc[:, 1::2]
X_test = normalized_test_c.iloc[:, 1::2]
```

```python
X_train = X_train.T.reset_index(drop=True)
```

```python
X_test = X_test.T.reset_index(drop=True)
```

```python
Y_train = [ (0 if i < 47 else 1)  for i in range(445)]
```

```python
Y_test = [ (0 if i < 50 else 1)  for i in range(450)]
```

```python
print(f"X_train: {X_train.shape}")
print(f"X_test: {X_test.shape}")
print(f"y_train: {Y_train.__len__()}")
print(f"y_test: {Y_test.__len__()}")
```

```python
# Create an SVM classifier
svm_classifier = SVC(kernel='linear') 

# Train the SVM classifier
svm_classifier.fit(X_train, Y_train)

# Make predictions on the test set
y_pred = svm_classifier.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(Y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
