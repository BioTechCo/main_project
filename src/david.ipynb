# Analysis of DAVID result 

```python
import pandas as pd
data = pd.read_csv('../result/logFC_pathway.csv', sep=',')
```

```python
hyper = pd.read_csv('../result/DMP_logFC.csv')
hyper = hyper[['gene','ID','dBeta','feat.cgi','logFC']]
hyper = hyper.sort_values(by=['dBeta'], ascending=False)
hyper = hyper[hyper['logFC'] > 0.35]
hyper
```

```python
import pandas as pd

gene_term_mapping = {}
for _, row in data.iterrows():
    genes = row['Genes'].split(', ')
    term = row['Term']
    for gene in genes:
        if gene not in gene_term_mapping:
            gene_term_mapping[gene] = []
        gene_term_mapping[gene].append(term)
def map_genes_to_pathways(gene):
    return gene_term_mapping.get(gene, '')

hyper['pathways'] = hyper['gene'].apply(map_genes_to_pathways)
hyper.reset_index(inplace=True)

hyper.head(20)
```

```python
hyper['index'] = hyper.index + 1
hyper
```

```python
hyper.to_csv('../result/logFC_hyper.csv', index=False)
```
