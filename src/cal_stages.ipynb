```python
from itertools import product

def count_methylation_patterns(dataframe, columns):
    # Get all unique values in the specified columns
    unique_values = dataframe[columns].stack().unique()

    # Generate all possible combinations of values (8 patterns)
    all_patterns = list(product(unique_values, repeat=len(columns)))

    # Count occurrences of each pattern
    pattern_counts = {
        pattern: dataframe[(dataframe[columns[0]] == pattern[0]) &
                           (dataframe[columns[1]] == pattern[1]) &
                           (dataframe[columns[2]] == pattern[2])].shape[0]
        for pattern in all_patterns
    }

    return pattern_counts
```
