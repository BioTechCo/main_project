```python
import pandas as pd
df_dbeta = pd.read_csv("../breast/result/GSE243529/train50/dbeta_hyper_tss_0.01.csv")
# result_max_per_gene_single_GSE243529_filter001_hyper_TSS
df_dbeta
```

```python
import pandas as pd
df_beta_train = pd.read_csv("../breast/champ_result/GSE243529/all_beta_normalized_0.csv")
df_beta_train
```

```python
df_mix_train = df_beta_train
```

## 重新分配列編號

```python
columns_to_drop = [str(i) for i in range(1, 525) if i % 2 == 0]
train = df_mix_train.drop(columns=columns_to_drop)


new_columns = ['Unnamed: 0'] + [str(i) for i in range(1, len(train.columns))]
train.columns = new_columns

train.set_index('Unnamed: 0', inplace=True)
train_X = train.T
train_y = [ (0 if i > 128 else 1)  for i in range(len(train.columns))]


print(train_X)
```

```python
df_beta_test = pd.read_csv("../breast/champ_result/GSE243529/all_beta_normalized_1.csv")
df_beta_test
```

```python

df_mix_test = pd.merge( left= df_dbeta,right=df_beta_test,on='Unnamed: 0')
df_mix_test
```

## 確認缺失的列標籤

```python
df_mix_test.columns = df_mix_test.columns.astype(str)

# Get the existing column labels.
existing_columns = set(df_mix_test.columns)

# Find the numbers between 0 and 525 that are not in the existing_columns
missing_columns = [str(i) for i in range(1,525) if str(i) not in existing_columns]

print("Missing columns from 0 to 525:")
print(missing_columns)
```

```python
columns_to_drop = [str(i) for i in range(1, 525) if (i % 2 == 0) and (str(i) not in missing_columns)]

test = df_mix_test.drop(columns=columns_to_drop)

new_columns = ['Unnamed: 0','dbeta', 'gene'] + [str(i) for i in range(1, len(test.columns)-2 )]
test.columns = new_columns

test = test.drop(columns=['dbeta', 'gene'])

test.set_index('Unnamed: 0', inplace=True)
test_X = test.T

test_y = [ (0 if i > 127 else 1)  for i in range(len(test.columns))]


print(test_X)
```

## 分配train/dev/test 

```python

from sklearn.model_selection import train_test_split

X = pd.concat([train_X, test_X], axis=0, ignore_index=True)
y = train_y + test_y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

X_train,X_dev,y_train,y_dev = train_test_split(X_train, y_train, test_size=0.3, random_state=42)

# train/dev/test
# 0.525/0.225/0.25

 
# X_train = train_X
# X_test = test_X
# y_train = train_y
# y_test = test_y
```

## 選取要預測的位點

```python
cgList = ['cg19935065', 'cg16201674', 'cg02355304']

X_train_select =  X_train[cgList]
y_train_select = y_train

X_test_select = X_test[cgList]
y_test_select = y_test

X_dev_select = X_dev[cgList]
y_dev_select = y_dev
```

## Random forest  

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error, accuracy_score, f1_score, classification_report
from scipy.optimize import dual_annealing
import xgboost as xgb


# Define the objective function
def objective(params):
    n_estimators = int(params[0])
    max_depth = int(params[1])
    min_samples_split = int(params[2])
    min_samples_leaf = int(params[3])
    
    model = RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        random_state=42
    )
    
    model.fit(X_train_select, y_train_select)
    predictions = model.predict(X_dev_select)
    mse = mean_squared_error(y_dev_select, predictions)
    return mse

# Set the parameter bounds for Simulated Annealing
bounds = [
    (10, 200),       # n_estimators
    (1, 50),         # max_depth
    (2, 20),         # min_samples_split
    (1, 20)          # min_samples_leaf
]

# Run Simulated Annealing optimization
result = dual_annealing(objective, bounds, maxiter=100)

# Best parameters
best_params = result.x
print(f"Best parameters: n_estimators={int(best_params[0])}, max_depth={int(best_params[1])}, min_samples_split={int(best_params[2])}, min_samples_leaf={int(best_params[3])}")

# Retrain the model with the best parameters
best_model = RandomForestRegressor(
    n_estimators=int(best_params[0]),
    max_depth=int(best_params[1]),
    min_samples_split=int(best_params[2]),
    min_samples_leaf=int(best_params[3]),
    random_state=42
)
best_model.fit(X_train_select, y_train_select)
predictions = best_model.predict(X_test_select)
mse = mean_squared_error(y_test_select, predictions)
print(f"Mean Squared Error on test set: {mse}")
```

## XGBoost

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error, accuracy_score, f1_score, classification_report, roc_auc_score, confusion_matrix, precision_score
from scipy.optimize import dual_annealing
import xgboost as xgb

# Define the objective function for regression using XGBoost
def objective(params):
    max_depth = int(params[0])
    learning_rate = params[1]
    n_estimators = int(params[2])
    min_child_weight = int(params[3])
    gamma = params[4]
    subsample = params[5]
    colsample_bytree = params[6]
    
    model = xgb.XGBRegressor(
        max_depth=max_depth,
        learning_rate=learning_rate,
        n_estimators=n_estimators,
        min_child_weight=min_child_weight,
        gamma=gamma,
        subsample=subsample,
        colsample_bytree=colsample_bytree,
        random_state=42,
        objective='reg:squarederror'
    )
    
    model.fit(X_train_select, y_train_select)
    predictions = model.predict(X_dev_select)
    mse = mean_squared_error(y_dev_select, predictions)
    return mse

# Set the parameter bounds for Simulated Annealing
bounds = [
    (1, 10),         # max_depth
    (0.01, 0.3),     # learning_rate
    (50, 500),       # n_estimators
    (1, 10),         # min_child_weight
    (0, 0.5),        # gamma
    (0.5, 1),        # subsample
    (0.5, 1)         # colsample_bytree
]

# Run Simulated Annealing optimization
result = dual_annealing(objective, bounds, maxiter=100)

# Best parameters
best_params = result.x
print(f"Best parameters: max_depth={int(best_params[0])}, learning_rate={best_params[1]}, n_estimators={int(best_params[2])}, min_child_weight={int(best_params[3])}, gamma={best_params[4]}, subsample={best_params[5]}, colsample_bytree={best_params[6]}")

# Retrain the model with the best parameters on the combined training and development set
best_model = xgb.XGBRegressor(
    max_depth=int(best_params[0]),
    learning_rate=best_params[1],
    n_estimators=int(best_params[2]),
    min_child_weight=int(best_params[3]),
    gamma=best_params[4],
    subsample=best_params[5],
    colsample_bytree=best_params[6],
    random_state=42,
    objective='reg:squarederror'
)
best_model.fit(X_train_select, y_train_select)
predictions = best_model.predict(X_test_select)
mse = mean_squared_error(y_test_select, predictions)
print(f"Mean Squared Error on test set: {mse}")
```
