```python
from sklearn.datasets import make_friedman1
from sklearn.feature_selection import RFECV
from sklearn.svm import SVR
from xgboost import XGBClassifier
import pandas as pd
```

資料引入(可更改)

```python
normalized_train = pd.read_csv('../result/all_beta_normalized_train_0.35.csv')
normalized_test = pd.read_csv('../result/all_beta_normalized_test_0.35.csv')


merged_df = pd.merge(normalized_train, normalized_test, on='Unnamed: 0')
merged_df = merged_df.iloc[:, 1::2]


y = [ (0 if (i < 47)or(i>444 and i <495 ) else 1)   for i in range(895)]
merged_array = merged_df.values.T
```

```python


estimator = XGBClassifier() 

# 5折交叉
selector = RFECV(estimator, step=1, cv=5)
selector = selector.fit(merged_array, y)
```

```python

print(selector.support_)

# 每个特征的得分排名，特征得分越低（1最好），表示特征越好
print(selector.ranking_)

#  挑选了几个特征
print(selector.n_features_)
# 每次交叉迭代各个特征得分
print(selector.cv_results_['mean_test_score'])


df = pd.DataFrame({
    'Probe_ID': normalized_train['Unnamed: 0'],
    'import_score_in_rfecv': selector.cv_results_['mean_test_score'],
    'import_rank_in_rfecv': selector.ranking_,
    'should_use': selector.support_
})


df.to_csv("check_rfecv_xgboost.csv",index=False)


selected_features = selector.support_

# 找到前三个特征的索引
selected_indices = [i for i, x in enumerate(selected_features) if x][:3]

# 输出选择的特征
print("Selected Features Indices: ", selected_indices)
```
