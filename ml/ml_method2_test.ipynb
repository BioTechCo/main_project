```python
import pandas as pd 
import numpy as np

df_beta_train = pd.read_csv("../GSE243529_RAW/all_beta_normalized_train.csv")
df_beta_train
```

```python
df_dbeta = pd.read_csv("../GSE243529_RAW/result_max_per_gene_single_GSE243529_filter001_hyper_TSS.csv")
df_dbeta
```

```python
df_mix_train = pd.merge( left= df_dbeta,right=df_beta_train,on='Unnamed: 0')
df_mix_train
```

```python
columns_to_drop = [str(i) for i in range(1, 525) if i % 2 == 0]
train = df_mix_train.drop(columns=columns_to_drop)

# # 重新編號樣本列
new_columns = ['Unnamed: 0', 'dbeta', 'gene'] + [str(i) for i in range(1, len(train.columns)-2 )]
train.columns = new_columns


train = train.drop(columns=['dbeta', 'gene'])
train.set_index('Unnamed: 0', inplace=True)
train_X = train.T
train_y = [ (0 if i > 128 else 1)  for i in range(len(train.columns))]


print(train_X)
```

```python
df_beta_test = pd.read_csv("../GSE243529_RAW/all_beta_normalized_test.csv")
df_beta_test
```

```python

df_mix_test = pd.merge( left= df_dbeta,right=df_beta_test,on='Unnamed: 0')
df_mix_test
```

```python
# df_mix_test.set_index('Unnamed: 0', inplace=True)
df_mix_test.columns = df_mix_test.columns.astype(str)

# 獲取現有的列標籤
existing_columns = set(df_mix_test.columns)

# 找出 0 到 525 之間不在 existing_columns 中的數字
missing_columns = [str(i) for i in range(1,525) if str(i) not in existing_columns]

# 打印缺失的列標籤
print("Missing columns from 0 to 525:")
print(missing_columns)
```

```python
columns_to_drop = [str(i) for i in range(1, 525) if (i % 2 == 0) and (str(i) not in missing_columns)]
# print(df_mix_test)
test = df_mix_test.drop(columns=columns_to_drop)

# print(df_mix_test)

# # 重新編號樣本列
new_columns = ['Unnamed: 0','dbeta', 'gene'] + [str(i) for i in range(1, len(test.columns)-2 )]
test.columns = new_columns

test = test.drop(columns=['dbeta', 'gene'])

test.set_index('Unnamed: 0', inplace=True)
test_X = test.T

test_y = [ (0 if i > 127 else 1)  for i in range(len(test.columns))]


print(test_X)
```

```python

from sklearn.model_selection import train_test_split

X = pd.concat([train_X, test_X], axis=0, ignore_index=True)
y = train_y + test_y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
# train/test
# 0.75/0.25

# X_train = train_X
# X_test = test_X
# y_train = train_y
# y_test = test_y

print(len(X_train),len(X_test))
print(X_train)
# X_train, X_dev, y_train, y_dev = train_test_split(X_train, y_train, test_size=0.3, random_state=42)
# train/dev/test
# 0.50/0.25/0.25
# print(len(X_train),len(X_dev),len(X_test))
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
from sklearn.model_selection import train_test_split

# 初始化隨機森林分類器
rf = RandomForestClassifier(n_jobs=-1, class_weight='balanced', max_depth=5, random_state=42)

# 初始化 Boruta
boruta = BorutaPy(rf, n_estimators='auto', verbose=2, random_state=42)

# 執行特徵選擇
# X_train.to_csv("t.csv")
# print(X_train.values)
boruta.fit(X_train.values, y_train)

# 獲取選中的特徵
X_selected = X_train.iloc[:, boruta.support_]

# 打印選中的特徵
print("Selected Features DataFrame:")
print(X_selected.head())

selected_features = X_train.columns[boruta.support_]
print("Selected Features: %s" % list(selected_features))

# 獲取特徵的排名
ranking = boruta.ranking_[boruta.support_]

# 將特徵及其排名組成 DataFrame
features_df = pd.DataFrame({
    'boruta_feature': selected_features,
})
```

```python
# 打印排序後的特徵名稱和排名
print("Sorted Selected Features with Rankings:")
print(features_df)
```

```python
import xgboost as xgb
from sklearn.metrics import classification_report


# 拆分數據集
X_train = X_train[features_df['boruta_feature']]
X_test = X_test[features_df['boruta_feature']]

clf = xgb.XGBClassifier(objective='binary:logistic', use_label_encoder=False, eval_metric='logloss')
clf.fit(X_train, y_train)

# 預測
y_pred = clf.predict(X_test)

# 打印分類報告
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFECV
from sklearn.metrics import classification_report
import shap

rf = RandomForestClassifier(n_jobs=-1, class_weight='balanced', max_depth=5, random_state=30)
rfecv = RFECV(estimator=rf, step=1, cv=5, scoring='accuracy')
rfecv.fit(X_train, y_train)

# 獲取 RFECV 選中的特徵
rfecv_features = X_train.columns[rfecv.support_]
print("RFECV Selected Features: %s" % list(rfecv_features))

rfecv_features_df=pd.DataFrame({
    'rfecv_feature': rfecv_features,
})
# 最終選擇的特徵
X_train_selected = X_train[rfecv_features]
X_test_selected = X_test[rfecv_features]

# 使用最終選擇的特徵重新訓練隨機森林模型
rf_final = RandomForestClassifier(n_jobs=-1, class_weight='balanced', max_depth=5, random_state=30)
rf_final.fit(X_train_selected, y_train)

# 預測
y_pred = rf_final.predict(X_test_selected)

# 打印分類報告
print(classification_report(y_test, y_pred))

```

```python
df_cluster = pd.read_csv("../GSE243529_RAW/Gene_Clusters_0.01_3cluster.csv")
df_cluster
```

```python
df_cluster_dbeta = pd.merge(left= df_cluster,right =df_dbeta,left_on ='Gene',right_on='gene' )
df_cluster_dbeta = df_cluster_dbeta.drop(columns=['gene'])
print(df_cluster_dbeta)

df_cluster_select = pd.merge(left= df_cluster_dbeta,right =inner,left_on ='Unnamed: 0',right_on='feature' )
df_cluster_select=df_cluster_select.drop(columns=['feature'])
print(df_cluster_select)
```

```python
# 根據 Cluster 分組
import pandas as pd
import itertools

grouped = df_cluster_select.groupby('Cluster')

# 生成排列組合
combinations = list(itertools.product(*[group.values.tolist() for _, group in grouped]))

# 將結果轉換為所需格式
result = []
for combo in combinations:
    result.append([row[2] for row in combo])  # row[0] 對應的是 Gene 列

# 打印結果
for combo in result:
    print(combo)
```

```python

from sklearn.metrics import classification_report, roc_auc_score,confusion_matrix

clf = RandomForestClassifier(n_jobs=-1, class_weight='balanced', max_depth=5, random_state=22)

with open('../GSE243529_RAW/classification_report_TSS_hyper001.txt', 'w') as f:
    for combo in result:
        
        X_train_c = X_train[combo]
        X_test_c = X_test[combo]
        
        clf.fit(X_train_c, y_train)
        
        y_pred = clf.predict(X_test_c)
        y_pred_prob = clf.predict_proba(X_test_c)[:, 1]
        
        report = classification_report(y_test, y_pred)
        auc = roc_auc_score(y_test, y_pred_prob)
        
        # 計算混淆矩陣
        cm = confusion_matrix(y_test, y_pred)

        # 提取 TN, FP, FN, TP
        tn, fp, fn, tp = cm.ravel()

        # 計算敏感性和特異性
        sensitivity = tp / (tp + fn)
        specificity = tn / (tn + fp)
        
        f.write(f"Combination: {combo}\n")
        f.write(f"AUC: {auc}\n")
        f.write(report)
        f.write(f"sensitivity: {sensitivity}  specificity:{specificity} \n")
        f.write("\n" + "="*80 + "\n")

print("Classification reports saved to classification_report.txt")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, roc_auc_score

# 示例的特征组合
combo = ['cg25869317', 'cg27298252', 'cg13379236']
X_train_c = X_train[combo]
X_test_c = X_test[combo]

# 定义参数网格
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# 初始化随机森林分类器
clf = RandomForestClassifier()

# 初始化网格搜索
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, scoring='roc_auc', cv=5, verbose=2, n_jobs=-1)

# 执行网格搜索
grid_search.fit(X_train_c, y_train)

# 输出最佳参数
print("Best parameters found: ", grid_search.best_params_)

# 使用最佳参数进行预测
best_clf = grid_search.best_estimator_
y_pred = best_clf.predict(X_test_c)
y_pred_prob = best_clf.predict_proba(X_test_c)[:, 1]

# 计算分类报告和 AUC
report = classification_report(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred_prob)

# 打印分类报告和 AUC
print("Classification Report:\n", report)
print("AUC Score:", auc)
```
