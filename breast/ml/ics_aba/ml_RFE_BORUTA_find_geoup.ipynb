```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate

# 初始化隨機森林模型
random_forest = RandomForestClassifier(random_state=42)

# 手動調整隨機森林模型參數
random_forest.set_params(n_estimators=120, max_depth=3, min_samples_split=2, min_samples_leaf=8)
# (n_estimators=150, max_depth=3, min_samples_split=2, min_samples_leaf=8)
X_train_try_param = X_train[clustering]


cv_results = cross_validate(random_forest, X_train_try_param, y_train, cv=5, scoring='accuracy', return_train_score=True)

# 輸出調整後的訓練集和測試集的交叉驗證平均準確率
print(f"Train accuracy (Random Forest): {cv_results['train_score'].mean():.4f}")
print(f"Test accuracy (Random Forest): {cv_results['test_score'].mean():.4f}")
```

```python

from sklearn.linear_model import LogisticRegression

logistic_regression = LogisticRegression(random_state=42, max_iter=300)
logistic_regression.set_params(C=1000, penalty='l2', solver='liblinear')
cv_results_lr = cross_validate(logistic_regression, X_train_try_param, y_train, cv=5, scoring='accuracy', return_train_score=True)
print(f"Train accuracy (Logistic Regression): {cv_results_lr['train_score'].mean():.4f}")
print(f"Test accuracy (Logistic Regression): {cv_results_lr['test_score'].mean():.4f}")
```

```python

from sklearn.svm import SVC

svm = SVC(random_state=42, probability=True)
svm.set_params(C=10, kernel='rbf', gamma='scale')
cv_results_svm = cross_validate(svm, X_train_try_param, y_train, cv=5, scoring='accuracy', return_train_score=True)
print(f"Train accuracy (SVM): {cv_results_svm['train_score'].mean():.4f}")
print(f"Test accuracy (SVM): {cv_results_svm['test_score'].mean():.4f}")
```

```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from skopt import BayesSearchCV
from skopt.space import Categorical, Real


tree_params = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}

clf_tree = DecisionTreeClassifier()

lr_params = [
    {'penalty': ['l1', 'l2'], 'C': [0.1,1, 10, 100], 'solver': ['liblinear'], 'max_iter': [100, 200, 300]},
    {'penalty': ['elasticnet'], 'C': [ 0.1,1, 10, 100], 'solver': ['saga'], 'max_iter': [100, 200, 300], 'l1_ratio': [0, 0.5, 1]},
    {'penalty': ['l1', 'l2'], 'C': [0.1,1, 10, 100], 'solver': ['saga'], 'max_iter': [100, 200, 300]},
    {'penalty': [None], 'solver': ['saga'], 'max_iter': [100, 200, 300]}
]

clf_lr = LogisticRegression(max_iter=500) 


rf_params = {
    'n_estimators': Categorical([100, 130,170, 200]),  # 使用 Categorical 表示離散值
    'max_depth': Categorical([1, 3, 5]),
    'min_samples_split': Categorical([2, 5, 10]),
    'min_samples_leaf': Categorical([2, 5, 10]),
    'max_features': Categorical(['sqrt', 'log2', 0.2, 0.5]),
    'bootstrap': [True],  # 這個值是布林值，不需要變更
    'max_samples': Categorical([0.6, 0.8])  # 明確為 Categorical
}
clf_rf = RandomForestClassifier()

svm_params = {
    'C': [1, 10, 100],
    'gamma': ['scale', 'auto'],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
}

clf_svm = SVC(probability=True)


xgb_params = {
    'n_estimators': Categorical([10, 15, 20]),
    'learning_rate': Real(0.01, 0.05),
    'max_depth': Categorical([2, 3, 5]),
    'subsample': Categorical([0.6, 0.8]), 
    'colsample_bytree': Categorical([0.6, 0.8]), 
    'gamma': Real(0.1, 0.3),
    'lambda': Real(1, 1.5, prior='uniform'), 
    'alpha': Real(0, 0.1, prior='uniform') 
}

clf_xgb = XGBClassifier( eval_metric='logloss')

results = []

model_list = [clf_tree,clf_lr,clf_rf,clf_svm,clf_xgb]
model_param = [tree_params,lr_params,rf_params,svm_params,xgb_params]
model_names = ['Decision Tree', 'Logistic Regression', 'Random Forest', 'SVM', 'XGBoost']

for cg in cg_list:
    X_train_c = X_train[cg]
    X_test_c = X_test[cg]
    
    for model,param, model_name in zip(model_list,model_param, model_names):
        
        model = BayesSearchCV(
            estimator=model, 
            search_spaces=param, 
            n_iter=10,  # 優化的疊代次數，可以調整
            cv=5,       # 5 折交叉驗證
            n_jobs=-1, 
            verbose=0, 
            random_state=42,  # 設定隨機種子
            scoring='f1'
        )
        model.fit(X_train_c, y_train)
        
        y_pred = model.predict(X_test_c)
        y_pred_prob = model.predict_proba(X_test_c)[:, 1]
        
        y_pred_train = model.predict(X_train_c)
        y_pred_prob_train = model.predict_proba(X_train_c)[:, 1]
        # report = classification_report(y_test, y_pred, output_dict=True)
        auc = roc_auc_score(y_test, y_pred_prob)
        
        train_accuracy = accuracy_score(y_train, y_pred_train)
        test_accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        sensitivity = recall_score(y_test, y_pred)
        specificity = recall_score(y_test, y_pred, pos_label=0)
        f1 = f1_score(y_test, y_pred)
        
        results.append( {
            'Model': model_name,
            'ID1': cg[0],
            'ID2': cg[1],
            'ID3': cg[2],
            # 'ID4': cg[3],
            'train_accuracy': train_accuracy,
            'test_accuracy': test_accuracy,
            'train - test accuracy' : train_accuracy - test_accuracy,
            'AUC': auc,
            'f1-score': f1,
            'sensitivity': sensitivity,
            'specificity': specificity,
            'precision': precision,
            'sen+spe': sensitivity + specificity,
            'f1+pre': f1 + precision,
            'AVG_(sen+spe)': (sensitivity + specificity)/2,
            'AVG_(f1+pre)': (f1 + precision)/2,
            'AVG_(avg(sen+spe)+avg(f1+pre))': ((f1 + precision)/2+ (sensitivity + specificity)/2)/2
        })
df = pd.DataFrame(results)

df

# df.to_csv("../../result/GSE243529/ics_mthod1_0706/train75/test_with_5model_0904_test_rc.csv",index=False)
```
