```python
import pandas as pd 
import numpy as np

df_beta_train = pd.read_csv("result/GSE243529_aba/X_train_sorted_0.8.csv")
df_beta_train
```

```python
df_dbeta = pd.read_csv("result/GSE243529_aba/dbeta_0.8_abs_0.02_hyper_TSS.csv")
# result_max_per_gene_single_GSE243529_filter001_hyper_TSS
df_dbeta
```

```python
df_mix_train = pd.merge( left= df_dbeta,right=df_beta_train,left_on= 'Unnamed: 0',right_on='Unnamed: 0')
df_mix_train
```

```python

from sklearn.model_selection import train_test_split


X_train = df_mix_train.drop(columns=["Unnamed: 0","dbeta","gene"])
X_train = X_train.T

X_train_normal_count = 210

if (X_train_normal_count):
    y_train = [(0 if i < (X_train_normal_count) else 1) for i in range(len(X_train))]
    
X_train = X_train.T


X_train = pd.concat([ df_mix_train["Unnamed: 0"], X_train], axis=1)
X_train.set_index("Unnamed: 0", inplace=True)

X_train = X_train.T

X_train
print(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate

# 初始化隨機森林模型
random_forest = RandomForestClassifier(random_state=42)

# 手動調整隨機森林模型參數
random_forest.set_params(n_estimators=155, max_depth=2, min_samples_split=2, min_samples_leaf=35)
# 180/2/2/35
# 150/2/2/35
# 使用交叉驗證，並同時返回訓練集和測試集的準確率
cv_results = cross_validate(random_forest, X_train, y_train, cv=5, scoring='accuracy', return_train_score=True)

# 輸出調整後的訓練集和測試集的交叉驗證平均準確率
print(f"Train accuracy (Random Forest): {cv_results['train_score'].mean():.4f}")
print(f"Test accuracy (Random Forest): {cv_results['test_score'].mean():.4f}")
```

```python
from scipy.optimize import dual_annealing
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np

# 定義目標函數，該函數通過5折交叉驗證來評估每個參數組合的性能
def objective(params):
    # 將參數轉換為整數
    n_estimators = int(params[0])
    max_depth = int(params[1])
    min_samples_split = int(params[2])
    min_samples_leaf = int(params[3])
    bootstrap = bool(params[4])

    clf = RandomForestClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        bootstrap=bootstrap,
        n_jobs=-1,
        random_state=42
    )

    # 使用5折交叉驗證來評估模型的F1分數，並返回負的平均分數（因為我們希望最小化）
    score = -cross_val_score(clf, X_train, y_train, cv=5, scoring='f1').mean()
    return score

# 定義參數空間的範圍
param_bounds = [
    (145, 185),   # n_estimators 的範圍，避免過多的樹
    (2, 4),       # max_depth 的範圍，進一步限制樹的最大深度
    (2, 7),      # min_samples_split 的範圍
    (20, 40),       # min_samples_leaf 的範圍
    (0, 1)        # bootstrap (0 表示 False，1 表示 True)
]

# 使用 dual_annealing 進行模擬退火優化
result = dual_annealing(
    objective,     # 目標函數
    param_bounds,  # 參數空間
    maxiter=30,   # 模擬退火的最大迭代次數
    seed=42
)

# 最佳結果
print("最佳參數組合:", result.x)
print("最佳目標值 (最小化的F1損失):", result.fun)
```

```python
best_rf = RandomForestClassifier(
    n_estimators=int(result.x[0]),  # 需要將浮點數轉換為整數
    max_depth=int(result.x[1]),
    min_samples_split=int(result.x[2]),
    min_samples_leaf=int(result.x[3]),
    bootstrap=bool(result.x[4]),  # 需要將0或1轉換為布爾值
    n_jobs=-1,
    random_state=42
)

best_rf
```

```python
# 用最佳參數再次進行5折交叉驗證，這次用來評估最終結果
from sklearn.model_selection import cross_val_predict

cv_scores = cross_val_score(best_rf, X_train, y_train, cv=5, scoring='f1')
predictions = cross_val_predict(best_rf, X_train, y_train, cv=5)

# 打印五折交叉驗證結果和預測
print(f'5-fold CV accuracy scores with best hyperparameters: {cv_scores}')
print(f'f1: {np.mean(cv_scores)}')
# print(f'Predictions from 5-fold CV with best hyperparameters: {predictions}')
```

```python
from sklearn.model_selection import cross_validate
# 使用 cross_validate 來同時獲取訓練和測試的準確率
cv_results = cross_validate(best_rf, X_train, y_train, cv=5, scoring='f1', return_train_score=True)

# 取得訓練準確率和測試準確率
train_accuracy = cv_results['train_score']
test_accuracy = cv_results['test_score']

# 分別輸出訓練準確率和測試準確率的平均值
print("Average train accuracy:", train_accuracy.mean())
print("Average test accuracy:", test_accuracy.mean())
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
from sklearn.model_selection import train_test_split

# 初始化隨機森林分類器
rf = best_rf

# 初始化 Boruta
boruta = BorutaPy(rf, n_estimators='auto', verbose=2, random_state=42)

# 執行特徵選擇
# X_train.to_csv("t.csv")
# print(X_train.values)

y_train = np.array(y_train)
boruta.fit(X_train.values, y_train)

# 獲取選中的特徵
X_selected = X_train.iloc[:, boruta.support_]

# 打印選中的特徵
print("Selected Features DataFrame:")
print(X_selected.head())

selected_features = X_train.columns[boruta.support_]
print("Selected Features: %s" % list(selected_features))

# 獲取特徵的排名
ranking = boruta.ranking_[boruta.support_]

# 將特徵及其排名組成 DataFrame
features_df = pd.DataFrame({
    'boruta_feature': selected_features,
})
```

```python
# 打印排序後的特徵名稱和排名
print("Sorted Selected Features with Rankings:")
print(features_df)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFECV
from sklearn.metrics import classification_report,make_scorer,f1_score


rf = best_rf

scoror = make_scorer(f1_score, average='macro')

rfecv = RFECV(estimator=rf, step=1, cv=5, scoring=scoror)
# rfecv = RFECV(estimator=rf, step=1, cv=5, scoring='accuracy')
rfecv.fit(X_train, y_train)

# 獲取 RFECV 選中的特徵
rfecv_features = X_train.columns[rfecv.support_]
print("RFECV Selected Features: %s" % list(rfecv_features))

rfecv_features_df=pd.DataFrame({
    'rfecv_feature': rfecv_features,
})
```
