```python
# X1 = train_normal_0.T + train_normal_1.T
X1 = pd.concat([train_normal_0, train_tumor_0], axis=0)


if len(train_normal_0):
    y1 = [(0 if i < len(train_normal_0) else 1) for i in range(len(X1))]
# train : 128t/134n = 262  ⇒ 256/268 = 524
# test : 127t/134n = 262 ⇒ 254/268 = 522

print((X1))
```

```python
def IQR(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    upper_fence = Q3 + IQR*1.5
    lower_fence = Q1 - IQR*1.5
    return upper_fence,lower_fence
def no_outlier(df):
    upper_fence, lower_fence = IQR(df)
    ddf=df[(df>lower_fence)&(df<upper_fence)]
    return ddf
```
