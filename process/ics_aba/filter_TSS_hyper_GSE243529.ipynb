```python
import pandas as pd

df_gene_single_dbeta = pd.read_csv("result/GSE243529_aba/dbeta.csv")
df_gene_single_dbeta
```

```python
df_gene_single_dbeta_threshold = df_gene_single_dbeta[abs((df_gene_single_dbeta['dbeta'])) > 0.02]
print(df_gene_single_dbeta_threshold,df_gene_single_dbeta_threshold.shape)


df_gene_single_dbeta_threshold_hyper = df_gene_single_dbeta_threshold[df_gene_single_dbeta_threshold['dbeta']>0]
print(df_gene_single_dbeta_threshold_hyper,df_gene_single_dbeta_threshold_hyper.shape)


df_gene_single_dbeta_threshold_hyper_TSS = df_gene_single_dbeta_threshold_hyper.loc[(df_gene_single_dbeta_threshold_hyper['feature'] == 'TSS200') | (df_gene_single_dbeta_threshold_hyper['feature'] == 'TSS1500')]
print(df_gene_single_dbeta_threshold_hyper_TSS,df_gene_single_dbeta_threshold_hyper_TSS.shape)
```

```python
df_gene_single_dbeta_threshold_hyper_TSS = df_gene_single_dbeta_threshold_hyper_TSS[["Unnamed: 0","dbeta","gene"]]
df_gene_single_dbeta_threshold_hyper_TSS.to_csv("result/GSE243529_aba/dbeta_0.8_abs_0.02_hyper_TSS.csv",index=False)
```
