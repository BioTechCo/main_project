#### 450K 組織計算hyper/hypo

```python
import pandas as pd
df_filter =pd.read_csv('../../資料整理/3_train_dbeta_0.35.csv')
df_filter
```

```python
train=pd.read_csv('../../all_beta_normalized_train.csv')
```

```python
df_gene = train[train['Unnamed: 0'].isin(df_filter['ID'])]
df_gene.reset_index(drop=True, inplace=True)
print(df_gene.shape)
df_gene.head()
```

```python
train_normal = df_gene.iloc[:,1:94:2]
train_tumor = df_gene.iloc[:,94::2]

train_tumor.columns=range(0,train_tumor.shape[1])
train_normal.columns=range(0,train_normal.shape[1])

train_tumor = train_tumor.T
train_normal = train_normal.T 
```

```python
train_normal_avg = train_normal.mean(skipna=True, axis=0)
tt_dbeta=(train_tumor).subtract(train_normal_avg, axis = 1)
```

```python
tt_dbeta #tumor delta beta
```

標註Hyper/Hypo

```python
avg_tumordBeta = tt_dbeta.mean(axis=0) #avg( Δβ (tumor) ) 
avg_tumordBeta
```

```python
methylation_level_df = pd.DataFrame({'ID': df_gene['Unnamed: 0'], 'methylation_level': avg_tumordBeta.apply(lambda x: 'Hyper' if x > 0 else 'Hypo')})
methylation_level_df
```

```python
cp_data=pd.read_csv('../result/result_basic_dbeta_0.35.csv')
cp_data.rename(columns={'Unnamed: 0': 'ID'}, inplace=True)
cp_data
```

```python
merged_df = pd.merge(cp_data, methylation_level_df, on='ID', how='inner')
merged_df
```

計算Up/Down regulation(使用F1 cutpoint為基準)

```python
cp=pd.read_csv("../../資料整理/1_388gene_cutpoint.csv")
cp
```

```python
import numpy as np
cut_point = cp["F1 cut-point"]

regulation = pd.DataFrame({
    'ID': df_gene['Unnamed: 0'],
    'regulation': np.where(avg_tumordBeta > cut_point, 'Up', 'Down')
})
regulation
```

```python
merged_df=pd.merge(merged_df, regulation, on='ID', how='inner')
merged_df
```

```python
merged_df.to_csv("../result/result_regulation_dbeta_0.35.csv",index=False)
```

#### GSE89093 液態計算hyper/hypo

```python
#GSE89093(46Normal/46Tumor)
import pandas as pd
gene_dbeta=pd.read_csv("../result/result_basic_dbeta_GSE89093.csv")
gene_dbeta=gene_dbeta[((abs(gene_dbeta['dbeta'])>0.03)&((gene_dbeta['UCSC_RefGene_Group']=='TSS1500')|(gene_dbeta['UCSC_RefGene_Group']=='TSS200')))]
gene_dbeta
```

```python
liquid_GSE89093=pd.read_csv("../../乳癌液態/GSE89093/GSE89093_all_beta_normalized.csv")
liquid_label_GSE89093=pd.read_csv("../../乳癌液態/GSE89093/GSE89093_phenotype.csv")
liquid_label_GSE89093=pd.DataFrame(liquid_label_GSE89093["cancer_status"])
liquid_GSE89093= liquid_GSE89093[liquid_GSE89093['ID_REF'].isin(gene_dbeta['ID'])]
liquid_GSE89093
```

```python
liquid_normal_GSE89093 = []
liquid_tumor_GSE89093 = []
for i in range(liquid_label_GSE89093.shape[0]):
    if liquid_label_GSE89093.iloc[i,0] == 'healthy':
        liquid_normal_GSE89093.append(liquid_GSE89093.iloc[:,i+1])
    else:
        liquid_tumor_GSE89093.append(liquid_GSE89093.iloc[:,i+1])
        
liquid_normal_GSE89093=pd.DataFrame(liquid_normal_GSE89093).T
liquid_tumor_GSE89093=pd.DataFrame(liquid_tumor_GSE89093).T
liquid_normal_GSE89093 = liquid_normal_GSE89093.astype(float)
liquid_tumor_GSE89093 = liquid_tumor_GSE89093.astype(float)
liquid_tumor_GSE89093=pd.concat([liquid_GSE89093.iloc[:, 0],liquid_tumor_GSE89093], axis=1)
liquid_normal_GSE89093=pd.concat([liquid_GSE89093.iloc[:, 0],liquid_normal_GSE89093], axis=1)
```

```python
train_normal_avg = liquid_normal_GSE89093.iloc[:,1::].mean(skipna=True, axis=1)
tt_dbeta=(liquid_tumor_GSE89093.iloc[:,1::]).subtract(train_normal_avg, axis = 0)
tt_dbeta
```

```python
avg_tumordBeta = tt_dbeta.mean(axis=1) #avg( Δβ (tumor) ) 
methylation_level_df = pd.DataFrame({'ID': liquid_GSE89093['ID_REF'], 'methylation_level': avg_tumordBeta.apply(lambda x: 'Hyper' if x > 0 else 'Hypo')})
methylation_level_df
```

```python
merged_df = pd.merge(gene_dbeta, methylation_level_df, on='ID', how='inner')
merged_df.to_csv("../result/result_regulation_dbeta_0.03_liquid.csv",index=False)
```
