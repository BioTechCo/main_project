```python
import pandas as pd
```

### Parameter

```python
comorbidity = 'group'
```

### Settings

```python
source_path = f'../result/all_beta_normalized_result_test_{comorbidity}.csv'
method_list = ['J','CZ','ER','UI','F1']
```

```python
test_df = pd.read_csv(source_path)
print(test_df.shape)
test_df.head()
```

```python
test_df.columns
```

```python
for i in method_list:
    ascending = False
    if i == 'ER' or i == 'UI':
        ascending = True
    for j in method_list:    
        test_df[f'{i},{j}_rank'] = test_df[f'{i},{j}'].rank(ascending=ascending)
```

```python
DMP_result = pd.read_csv('../champ_result/DMP_result.csv')
DMP_result.rename(columns={'Unnamed: 0':'ID'}, inplace=True)
DMP = DMP_result.drop(DMP_result.columns[1:10], axis=1)
DMP = DMP.drop('gene', axis=1)
DMP.head()
```

```python
test_df = pd.merge(test_df, DMP, on='ID')
test_df.head()
```

```python
test_df.to_csv(f'../result/all_beta_normalized_result_test_{comorbidity}_rank.csv', index=False)
```
