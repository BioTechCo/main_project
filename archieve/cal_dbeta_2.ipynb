### Calculate delta beta

```python
import pandas as pd
df=pd.read_csv('C:/Users/Administrator/Documents/all_beta_normalized.csv')
df.rename(columns={'Unnamed: 0': 'ID'}, inplace=True)
```

```python
#[0~349419,1~94]>>normal
NM=df.iloc[0:,1:95]
NM=NM.T
NM=NM.drop_duplicates(keep='first')
#NM
```

```python
# Computing IQR
NM_Q1 = NM.quantile(0.25)
NM_Q3 = NM.quantile(0.75)
NM_IQR = NM_Q3 - NM_Q1
NM_upper_fence=NM_Q3+NM_IQR*1.5
NM_lower_fence=NM_Q1-NM_IQR*1.5
```

```python
print(NM_upper_fence)
print(NM_lower_fence)
```

```python
NM_with_no_outlier = NM[(NM>NM_lower_fence)&(NM<NM_upper_fence)]
NM_with_no_outlier
```

```python
#tumor
TM=df.iloc[0:,95:]
TM=TM.T
TM=TM.drop_duplicates(keep='first')
#TM
```

```python
# Computing IQR
TM_Q1 = TM.quantile(0.25)
TM_Q3 = TM.quantile(0.75)
TM_IQR = TM_Q3 - TM_Q1
TM_upper_fence=TM_Q3+TM_IQR*1.5
TM_lower_fence=TM_Q1-TM_IQR*1.5
```

```python
print(TM_upper_fence)
print(TM_lower_fence)
```

```python
TM_with_no_outlier = TM[(TM>TM_lower_fence)&(TM<TM_upper_fence)]
TM_with_no_outlier
```

```python
#data=pd.DataFrame(NM_with_no_outlier.T).join(TM_with_no_outlier.T)
#data
```

```python
#è¨ˆç®—normalå¹³å‡å€¼
NM_avg=NM_with_no_outlier.mean(skipna = True)
NM_avg
```

```python
#è¨ˆç®—delta beta
dSub=(TM_with_no_outlier.T).subtract(NM_avg, axis = 0)
dSub
```

```python
# Computing IQR
Q1 = dSub.T.quantile(0.25)
Q3 = dSub.T.quantile(0.75)
IQR = Q3 - Q1
upper_fence=Q3+IQR*1.5
lower_fence=Q1-IQR*1.5
print(upper_fence)
print(lower_fence)
```

```python
dSub1=dSub.T
dSub_with_no_outlier = dSub1[(dSub1>lower_fence)&(dSub1<upper_fence)]
dSub_with_no_outlier.head(20)
```

```python
dBeta=dSub_with_no_outlier.mean(skipna = True)
dBeta.head(20)
```

```python
dBeta=abs(dBeta)
dBeta
```
