### Use training data to find cut point, and use cut point to test data
### Identify the high/low methylation level/difference

```python
import pandas as pd
df = pd.read_csv('C:/Users/Administrator/Desktop/專題/all_beta_normalized_test.csv')
df.rename(columns={'Unnamed: 0': 'ID'}, inplace=True)
df
```

```python
NM = df.iloc[0:, 1:101]
NM=NM.T
NM=NM.drop_duplicates(keep='first')
NM
```

```python
#tumor
TM = df.iloc[0:, 101:]
TM = TM.T
TM = TM.drop_duplicates(keep='first')
TM
```

```python
#計算normal平均值
NM_avg = NM.mean(skipna=True)
NM_avg
```

```python
#計算delta beta>>Tumor
dBeta_TM = (TM.T).subtract(NM_avg, axis=0)
dBeta_TM
```

```python
dBeta_TM=dBeta_TM.T
dBeta = dBeta_TM.mean(skipna=True)
dBeta
```

```python
one = pd.DataFrame(dBeta, columns=['dBeta']).join(df['ID'])
one
```

```python
RS = pd.read_csv('C:/Users/Administrator/Desktop/專題/4/group/all_beta_normalized_result.csv')
#RS.rename(columns={'Unnamed: 0': 'ID'}, inplace=True)
RS
```

```python
#找ID交集
TN = pd.merge(one.iloc[0:, 1], RS, how='inner')
TN
#pd.DataFrame(TN)
#dBeta_TN = TN.dropna()
#dBeta_TN
```

```python
dBeta_TN1 = pd.merge(TN, one, how='inner')
dBeta_TN1
```

```python
RS1 = pd.read_csv('C:/Users/Administrator/Desktop/專題/4/group/test_result.csv')
#RS.rename(columns={'Unnamed: 0': 'ID'}, inplace=True)
RS1
```

```python
RS1.loc[RS1['dBeta'] > 0,'test_methylation_level'] = 'hyper'
RS1.loc[RS1['dBeta'] < 0, 'test_methylation_level'] = 'hypo'
RS1.loc[abs(RS1['dBeta']) > RS1['cut-point'],'test_methylation_difference'] = 'high'
RS1.loc[abs(RS1['dBeta']) < RS1['cut-point'],'test_methylation_difference'] = 'low'
RS1
```

```python
#Result
outputpath = 'C:/Users/Administrator/Desktop/專題/4/group/test_result.csv'
RS1.to_csv(outputpath, sep=',', index=False, header=True)
```
