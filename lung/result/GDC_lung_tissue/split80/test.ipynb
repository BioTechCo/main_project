```python
import pandas as pd
```

```python
df = pd.read_csv("../split20/all_beta_normalized.csv")
```

```python
df
```

```python
print("每列缺失值數量：")
print(df.isna().sum())

print("\nDataFrame 是否包含 NaN:", df.isna().values.any())
```

```python

# 1. 檢查每列的 NaN 值數量
print("每列缺失值數量：")
print(df.isna().sum())

# 2. 檢查整個 DataFrame 是否有 NaN
print("\nDataFrame 是否包含 NaN:", df.isna().values.any())

# 3. 顯示包含 NaN 的行
print("\n包含 NaN 的行：")
print(df[df.isna().any(axis=1)])
```

```python

# 4. 顯示 NaN 的比例
print("\n每列 NaN 值佔總行數的比例：")
print(df.isna().mean())

# 5. 視覺化 NaN 分佈（可選，需要安裝 `missingno`）
# pip install missingno
```

```python
dfn = df.dropna()
dfn
```

```python

# 1. 檢查每列的 NaN 值數量
print("每列缺失值數量：")
print(dfn.isna().sum())

# 2. 檢查整個 DataFrame 是否有 NaN
print("\nDataFrame 是否包含 NaN:", dfn.isna().values.any())

# 3. 顯示包含 NaN 的行
print("\n包含 NaN 的行：")
print(dfn[dfn.isna().any(axis=1)])
```

```python
dfn.to_csv("../split20/all_beta_normalized.csv",index=False)
```
